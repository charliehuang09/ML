```python
def sample(threshold, baseXP, extraXP):
    t = (1 - threshold) * 0.25 * baseXP
    a = baseXP * 5
    b = threshold * extraXP + (t * 1) + (t * 2) + (t * 3) + (t * 4)
    if (a >= b):
        out = "absorb"
    else:
        out = "craft"
    return a, b, out 
```

```python
#Ultra -> Super
sample(0.01, 50000000, 5000000000)
```
